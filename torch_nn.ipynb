```python
# advanced nn module with neurons

import torch
import torch.nn as nn
```

```python
class module(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.linear1=nn.Linear(num_features ,3)
        self.relu=nn.ReLU()
        self.linear2=nn.Linear(3,1)
        self.sigmoid=nn.Sigmoid()
    def forward(self , features):
        out=self.linear1(features)
        out=self.relu(out)
        out=self.linear2(out)
        out=self.sigmoid(out)
        return out
```

```python
features=torch.rand(10,5)
model1=module(features.shape[1])
model1(features)
```

```python
model1.linear1.weight
```

```python
summary(model1 , input_size=(10,5))
```

```python
# Sequential container

class module1(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.network =nn.Sequential(
            nn.Linear(num_features ,3),
            nn.ReLU(),
            nn.Linear(3,1),
            nn.Sigmoid(),
        )
    def forward(self , features):
        out=self.network(features)
        return out
```

```python
features=torch.rand(10,5)
model2=module1(features.shape[1])
model2(features)
```

```python

import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace=True, errors='ignore')
```

```python
x_train ,X_test ,y_train ,y_test=train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(x_train)
X_test = scaler.transform(X_test)
```

```python

encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
class MYSIMPLENN(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.linear=nn.Linear(num_features ,1)
        self.sigmoid=nn.Sigmoid()
    def forward(self ,x):
        out=self.linear(x)
        out=self.sigmoid(out)
        return out
    def loss_function(self, y_pred, y):
    # Clamp predictions to avoid log(0)
     epsilon = 1e-7
     y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)

    # Calculate loss
     loss = -(y_train_tensor * torch.log(y_pred) + (1 - y_train_tensor) * torch.log(1 - y_pred)).mean()
    
     return loss
    
```

```python
# important parameters
learing_rate=0.1
epoches=25
```

```python
num_features = X_train_tensor.shape[1]
```

```python
model=MYSIMPLENN(num_features)
```

```python
for epoch in range(epoches):

  # forward pass
  y_pred = model.forward()

  # loss calculate
  loss = model.loss_function(y_pred, y_train_tensor)

  # backward pass
  loss.backward()

  # parameters update
  with torch.no_grad():
    model.linear.weights -= learing_rate * model.weights.grad
    model.linear.bias -= learing_rate * model.bias.grad

  # zero gradients
  model.linear.weights.grad.zero_()
  model.linear.bias.grad.zero_()

  # print loss in each epoch
  print(f'Epoch: {epoch + 1}, Loss: {loss.item()}') 
```
