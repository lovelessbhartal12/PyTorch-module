```python
import torch
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=10, n_features=2 ,n_informative=2,n_redundant=0 ,n_classes=2, random_state=42)
```

```python
x=torch.tensor(x, dtype=torch.float32)
y=torch.tensor(y , dtype=torch.long) 
```

```python
y
```

```python
from torch.utils.data import Dataset, DataLoader
```

```python
class cusotomdataset(Dataset):

    def __init__(self , features ,lables):
        super().__init__()
        self.features=features
        self.labels=lables
    
    def __len__(self):
        return self.features.shape[0]
    
    def __getitem__(self, index):
        return self.features[index] ,self.labels[index]
```

```python
dataset=cusotomdataset(x,y)
```

```python
len(dataset)
```

```python
dataset[1]
```

```python
dataloader=DataLoader(dataset ,batch_size=2 , shuffle=True)
```

```python
for batch_features ,batch_lables in dataloader:
    print(batch_features)
    print(batch_lables)
    print("-"*50)
```

```python

import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace=True, errors='ignore')

```

```python
x_train ,X_test ,y_train ,y_test=train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(x_train)
X_test = scaler.transform(X_test)
```

```python

encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```
